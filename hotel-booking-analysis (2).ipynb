# ADR vs Reservation of Hotels 

# Resort Hotel

```python
# Resort Hotel
plt.figure(figsize=(30, 12))
resort_hotel=resort_hotel.groupby('Reservation Date')[['adr']].mean()
resort_hotel_filter = (resort_hotel.index > '2016') & (resort_hotel.index < '2017-09')
resort_hotel_filtered = resort_hotel[resort_hotel_filter].reset_index()
plt.plot(resort_hotel_filtered['Reservation Date'], resort_hotel_filtered['adr'], label='Resort Hotel')
plt.legend(fontsize=25)
plt.xlabel('Reservation Status Date',fontsize=20)
plt.ylabel('Average Daily Rate',fontsize=20)
plt.show()
```

# City Hotel

```python
# City Hotel
plt.figure(figsize=(30, 12))
city_hotel=city_hotel.groupby('Reservation Date')[['adr']].mean()
city_hotel_filter = (city_hotel.index > '2016') & (city_hotel.index < '2017-09')
city_hotel_filtered = city_hotel[city_hotel_filter].reset_index()
plt.plot(city_hotel_filtered['Reservation Date'], city_hotel_filtered['adr'], label='City Hotel')
plt.legend(fontsize=25)
plt.xlabel('Reservation Status Date',fontsize=20)
plt.ylabel('Average Daily Rate',fontsize=20)
plt.show()
```

```python
# Lets Combine Hotel vs City
plt.figure(figsize=(35, 16))
resort_hotel=resort_hotel.groupby('Reservation Date')[['adr']].mean()
city_hotel=city_hotel.groupby('Reservation Date')[['adr']].mean()
resort_hotel_filter = (resort_hotel.index > '2016') & (resort_hotel.index < '2017-09')
city_hotel_filter = (city_hotel.index > '2016') & (city_hotel.index < '2017-09')
resort_hotel_filtered = resort_hotel[resort_hotel_filter].reset_index()
city_hotel_filtered = city_hotel[city_hotel_filter].reset_index()
plt.plot(resort_hotel_filtered['Reservation Date'], resort_hotel_filtered['adr'], label='Resort Hotel')
plt.plot(city_hotel_filtered['Reservation Date'], city_hotel_filtered['adr'], label='City Hotel')
plt.legend(fontsize=25)
plt.xlabel('Reservation Status Date',fontsize=20)
plt.ylabel('Average Daily Rate',fontsize=20)
plt.show()
```

# Now Lets see month wise cancelation

```python
df['month']=df['Reservation Date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.title('Month wise Booking Status',size=20)
plt.legend(['not canceled','canceled'])
plt.show
```

# Now we will ADR wise months

# Cancelletion and Reservation with respect to ADR

```python
# For cancelled
cancelled_df_adr = cancelled_data.groupby('Reservation Date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('Reservation Date',inplace=True)

plt.figure(figsize=(30,15))
plt.plot(cancelled_df_adr['Reservation Date'],cancelled_df_adr['adr'],label='cancel')
plt.show()
```

```python
# For Not Cancelled
not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('Reservation Date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('Reservation Date',inplace=True)

plt.figure(figsize=(30,15))
plt.plot(not_cancelled_df_adr['Reservation Date'],not_cancelled_df_adr['adr'],label='not cancel')
plt.show()
```

```python
# lets combined cancelled with not cancelled
 
plt.figure(figsize=(30, 15))
plt.title('ADR To Cancelled & Not Cancelled',fontsize=20)

# Filter data based on conditions
cancelled_filter = (cancelled_df_adr['Reservation Date'] > '2016') & (cancelled_df_adr['Reservation Date'] < '2017-09')
not_cancelled_filter = (not_cancelled_df_adr['Reservation Date'] > '2016') & (not_cancelled_df_adr['Reservation Date'] < '2017-09')

# Plot filtered data
plt.plot(cancelled_df_adr.loc[cancelled_filter, 'Reservation Date'], cancelled_df_adr.loc[cancelled_filter, 'adr'], label='cancel')
plt.plot(not_cancelled_df_adr.loc[not_cancelled_filter, 'Reservation Date'], not_cancelled_df_adr.loc[not_cancelled_filter, 'adr'], label='not cancel')

plt.legend(fontsize=15)
plt.show()
```
